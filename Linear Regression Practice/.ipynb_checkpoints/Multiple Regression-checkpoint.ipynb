# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading Data

```python
real_estate = pd.read_csv("Real_estate.csv")
```

# Examine Data

```python
# Displaying the first 5 rows of table
real_estate.head()
```

```python
# Printing the columns
real_estate.columns
```

```python
# Removing the first column from the data set, not helpful
real_estate.drop(['No'], axis='columns', inplace = True)
```

```python
real_estate.head()
```

```python
# Renaming the columsn for easier implimentation
real_estate.rename({'X1 transaction date': 'TD', 'X2 house age': 'AGE', 'X3 distance to the nearest MRT station' : 'DTMRT', 'X4 number of convenience stores' : 'NCS', 'X5 latitude': 'LAT', 'X6 longitude': 'LONG', 'Y house price of unit area': 'PRICE'}, axis='columns', inplace=True)
```

```python
real_estate.head()
```

```python
# Printing the summary statistics of the data
real_estate.describe()
```

```python
# Display summary of the data frame item
real_estate.info()
```

```python
# Checking for null values
real_estate.isnull().sum(axis=0)
```

# Data Visualization

```python
# Setting the style and color of the plot
sns.set_palette("Blues_d")
sns.set_style("darkgrid")
```

```python
# Showing the distribution of the dependent variable Y (PRICE)
sns.distplot(real_estate['PRICE'])
```

```python
# Scatter plot showing the relationship of distance to public transport to price
sns.jointplot(x='DTMRT', y = 'PRICE', data=real_estate)
```

```python
# Adding a regression line to the plot
sns.lmplot(x='DTMRT', y='PRICE', data=real_estate)
```

```python
# Getting the plot for the relationship between every pair of variables
sns.pairplot(real_estate)
```
